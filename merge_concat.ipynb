```python
import pandas as pd 
```

```python
temp_df=pd.DataFrame({"city":['Hyderabad','Chennai', 'Mumbai','kolkatta', 'Banguluru'],
                      'temperature':[35,37,35,29,25],
                      'humidity':[23,56,67,34,65]})
                    
```

```python
temp_df
```

```python
wind_df=pd.DataFrame({'city':['Delhi','Hyderabad','Vijayawada','Mumbai','Jaipur'],
                      'wind':['23 km/hr','27 km/hr', '35 km/hr','25 km/hr','34 km/hr']}) 
```

```python
wind_df
```

```python
df1= pd.merge(temp_df, wind_df, on ='city')
df1
```

```python
df2= pd.merge(temp_df, wind_df, on = 'city', how= 'outer')
df2
```

```python
df3=pd.merge(temp_df,wind_df, on='city', how='left')
df3
```

# Concat

```python
df4= pd.concat([temp_df, wind_df])
df4
```

```python
df5= pd.concat([temp_df, wind_df], ignore_index= True)
df5
```

```python
df6= pd.concat([temp_df, wind_df], axis=1)
df6
```

## Group by

```python
path= "Glassdoor Gender Pay Gap.csv"
```

```python
df= pd.read_csv(path)
df.head(5)
```

```python
gdf1= df.groupby("JobTitle")["BasePay"].mean()
gdf1
```

```python
#gdf2= df.groupby("JobTitle")["BasePay","Bonus"].agg(['min','max','medain'])
gdf2= df.groupby("JobTitle")[["BasePay","Bonus"]].agg(["max", "min", "median"])
gdf2
```

```python
gdf3= df.groupby("JobTitle")["Gender"].value_counts()
gdf3
```

```python
gdf4= df.groupby("JobTitle")["Gender"].value_counts().unstack()
gdf4
```

```python
gdf5= df.groupby("JobTitle")[['Dept','Gender']].value_counts().unstack()
gdf5
```
